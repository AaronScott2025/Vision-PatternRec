```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
from skimage import io as io_url
```

```python
frame = io_url.imread('https://github.com/AaronScott2025/Vision-PatternRec/blob/main/Lena.png?raw=true')
frame = np.float32(frame) / 255
plt.imshow(frame)
```

```python
frame2 = io_url.imread('https://github.com/AaronScott2025/Vision-PatternRec/blob/main/baboon.png?raw=true')
frame2 = np.float32(frame2) / 255
plt.imshow(frame2)
```

```python
print(frame.shape)
print(frame2.shape)
```

```python
frm = frame[0:100,0:100,0]
plt.imshow(frm, cmap="gray")
```

```python
frm2 = frame2[0:100,0:100,0]
plt.imshow(frm2, cmap="gray")
```

```python
array = np.array(frame)

new_array = frm + 20
plt.imshow(new_array,cmap='gray')
```

```python
#Reference
frme = frame[100:400,100:400]
plt.imshow(frme)
```

```python
#Blur (Mean Filter)

new_array1 = frme
kernel = (1/9) * np.array([[1,1,1],[1,1,1],[1,1,1]])
new_array1 = cv2.filter2D(new_array1, -1, kernel)

plt.imshow(new_array1,cmap='gray')
```

```python
#Sharpening (Sharpen Filter)

new_array1 = frme
kernel = np.array([[0,0,0],[0,2,0],[0,0,0]])
result = kernel - (1/9) * np.array([[1,1,1],[1,1,1],[1,1,1]])
new_array1 = cv2.filter2D(new_array1, -1, result)

plt.imshow(new_array1,cmap='gray')
```

```python
edges = cv2.Canny(frame, 100, 200)

plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.show()
```
